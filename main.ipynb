```python
from dotenv import load_dotenv
load_dotenv()

import openai
import requests

client = openai.OpenAI()
```

```python
BASE_URL = "https://nomad-movies.nomadcoders.workers.dev"

TOOLS = [
    {
        "type": "function",
        "function": {
            "name": "get_popular_movies",
            "description": "현재 인기 있는 영화 목록을 가져옵니다.",
            "parameters": {"type": "object", "properties": {}, "required": []}
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_movie_details",
            "description": "특정 영화 ID의 상세 정보(제목, 줄거리, 평점, 개봉일)를 가져옵니다.",
            "parameters": {
                "type": "object",
                "properties": {
                    "id": {"type": "integer", "description": "영화 ID"}
                },
                "required": ["id"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_movie_credits",
            "description": "특정 영화 ID의 출연진 및 감독 정보를 가져옵니다.",
            "parameters": {
                "type": "object",
                "properties": {
                    "id": {"type": "integer", "description": "영화 ID"}
                },
                "required": ["id"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_similar_movies",
            "description": "특정 영화 ID와 비슷한 영화를 추천받습니다.",
            "parameters": {
                "type": "object",
                "properties": {
                    "id": {"type": "integer", "description": "영화 ID"}
                },
                "required": ["id"]
            }
        }
    },
]
```

```python
def get_popular_movies():
    r = requests.get(f"{BASE_URL}/movies")
    return r.json()

def get_movie_details(id):
    r = requests.get(f"{BASE_URL}/movies/{id}")
    return r.json()

def get_movie_credits(id):
    r = requests.get(f"{BASE_URL}/movies/{id}/credits")
    return r.json()

def get_similar_movies(id):
    r = requests.get(f"{BASE_URL}/movies/{id}/similar")
    return r.json()
```

```python
# Test 1: 인기 영화 조회
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "지금 인기 있는 영화가 무엇인지 알려줘"}],
    tools=TOOLS,
)

message = response.choices[0].message.tool_calls[0]
print(message)

result = get_popular_movies()
result
```

```python
# Test 2: 특정 영화 상세 정보
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "movie ID 550에 해당하는 영화가 무엇인지 알려줘"}],
    tools=TOOLS,
)

message = response.choices[0].message.tool_calls[0]
print(message)

result = get_movie_details(550)
result
```

```python
# Test 3: 특정 영화 출연진 조회
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "movie ID 550에 해당하는 영화에 누가 출연하는지 알려줘"}],
    tools=TOOLS,
)

message = response.choices[0].message.tool_calls[0]
print(message)

result = get_movie_credits(550)
result
```

```python
# Test 4: 유사 영화 조회
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "movie ID 550과 비슷한 영화들을 추천해줘"}],
    tools=TOOLS,
)

message = response.choices[0].message.tool_calls[0]
print(message)

result = get_similar_movies(550)
result
```
